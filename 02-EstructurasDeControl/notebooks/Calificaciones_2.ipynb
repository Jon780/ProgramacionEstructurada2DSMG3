## programa de calificaciones

este programa calcula promedio de calificaciones y evalua si un alumno Aprobo o No Aprovo

## definicion de variables 

```python
calif1 = 0.0
calif2 = 0.0
calif3 = 0.0
promedio = 0.0
```

## Entrada de datos

```python
calif1 = float(input("introduce la calificacion1"))
calif2 = float(input("introduce la calificacion2"))
calif3 = float(input("introduce la calificacion3"))
```

## Proceso: calculo de promedio

```python
promedio = (calif1 + calif2 + calif3) / 3
```

## Comparacion con una Estructura selectiva doble

```python
if promedio >= 7.0:
    print("Aprovado")
else:
    print("No Aprobado")
```
